# MooneyMood

**Can mood be induced by neutral stimuli, namely stimuli which resolve uncertainty?**

- Participants have to fill in the positive negative affect scale (PANAS) questionnaire before seeing the sequence of images. The PANAS is just a list of emotion words (10 positive and 10 negative) for which one has to indicate to what extent one has experienced each in the past week. (panas phase)
- A sequence of 20 Mooney images which morph in to their solution (reveal phase) (instruction: you will be asked to remember the objects in the images afterwards)
- Unannounced open question: produce as many items (emotion words) as possible from PANAS (memory phase). Some will be confabulated. This has been used as an indirect (non-self-report) measure of affect/mood (Hartig et al. 1999; Vuoskoski & Eerola, 2012). A form of mood congruent recall/confabulation.

Specifics:
- Two conditions: upright (resolving) and inverted (unresolving) Mooneys in phase 2
- 89 participants from Prolific, took 5-15min
- Each participant: random assignment to condition (upright or inverted) and random sample of 20 Mooneys from a pool of 835 images (better: best Mooneys)

**Hypothesis:
Participants in the upright condition will have a more positive mood (more positive/less negative words produced) than participants in the inverted condition, because it is pleasurable to see uncertainty resolve.**

```python

dffull = pd.DataFrame()

all_files = glob.glob(os.path.join('../data/', "*.csv")) # get list of individual data files

for f in all_files:
    print f
    df = pd.read_csv(f, index_col=0)
    df = df.set_index(df.trial_index)
    dffull = pd.concat([dffull,make_row(df)], ignore_index=True)

print 'Variables:\n', dffull.columns
print 'variables:\n', dffull.dtypes
print 'nb of participants:', len(dffull['subject'].unique())
print len(dffull)
# Save full data
dffull.to_csv('dffull.csv')
```

```python
import ast
from fuzzywuzzy import fuzz

def make_row(df):
    pos = ['proud','alert', 'active', 'determined', 'enthousiastic', 'attentive', 'strong', 'inspired', 'interested', 'excited']
    neg = ['distressed', 'upset', 'hostile', 'nervous', 'guilty', 'afraid', 'scared', 'ashamed', 'irritable', 'jittery']
    negadd = ['anxious', 'sad','unhappy', 'irrational', 'restless', 'tired', 'shame', 'hostility', 'annoyed', 'uneasy', 'jealous', 'angry','depressed', 'aggressive', 'fear', 'fearful', 'embarrassed', 'bored', 'worried', 'feared', 'afflicted', 'frustrated', 'frightened', 'hurt', 'stressed', 'agitated', 'emotional','sad', 'distracted','shy', 'irritated']
    posadd = ['motivated','impressed', 'optimistic','happiness', 'excitement', 'relaxing', 'happy', 'grateful','concentrated', 'imaginative', 'positive','accomplished', 'curious', 'achievement', 'ambitious', 'bold','contented', 'content', 'happy', 'elated', 'calm', 'energetic', 'joyful', 'relaxed', 'confident'] 

    reddf = df[df.test_part=="panas_questionnaire"]
    reddf['remembered'] = ast.literal_eval(df[df.test_part=="posttest"].responses.values[0])['Q0']
    reddf['totalTime']= reddf['totalTime']/60000

    if '.' in reddf['remembered'].values[0]:
        rememberedraw = reddf['remembered'].values[0].split('.');
    else: rememberedraw = reddf['remembered'].values[0].split(',');
    
    remembered = []
    for i in rememberedraw:
        res = i.split()
        res = [j.strip(' ,').strip().lower() for j in res]
        remembered= remembered + res

    poswords = 0
    negwords = 0
    confab = 0
    confabpos = 0
    confabneg = 0
    totalwords = 0

    for word in remembered:
        if word=='':
            continue
        totalwords +=1
        found = 0
        for i in pos:
            if fuzz.ratio(i,word) > 90:
                poswords+=1
                found = 1
                continue
        for i in neg:
            if fuzz.ratio(i,word) > 90:
                negwords+=1
                found = 1
                continue
        for i in posadd:
            if fuzz.ratio(i,word) > 90:
                poswords+=1
                confab+=1
                confabpos+=1
                found = 1
                continue
        for i in negadd:
            if fuzz.ratio(i,word) > 90:
                negwords+=1
                confab+=1
                confabneg+=1
                found = 1
                continue
        if not found: print word
    reddf['poswords'] = poswords
    reddf['negwords'] = negwords
    reddf['confabpos'] = confabpos
    reddf['confabneg'] = confabneg
    reddf['confab'] = confab
    reddf['rememberedwords'] = str(remembered)
    reddf['totalwords'] = totalwords
    #print remembered,poswords,negwords,confab,totalwords
    
    panasorder = reddf.order.values[0].split(',')
    panasresp = ast.literal_eval(reddf.responses.values[0])
    
    PAS = 0
    NAS = 0
       
    for i,v in enumerate(panasorder):
        if v.lower() in pos:
            PAS+= int(panasresp['Q'+str(i)])
        if v.lower() in neg:
            NAS+= int(panasresp['Q'+str(i)])
    #print NAS, PAS
    
    reddf['NAS']=NAS
    reddf['PAS']=PAS
    
    return reddf
```
